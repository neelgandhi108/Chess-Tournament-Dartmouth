```python
import chess


class ChessGame:
    def __init__(self, player1, player2):
        self.board = chess.Board()
        self.players = [player1, player2]

    def make_move(self):

        player = self.players[1 - int(self.board.turn)]
        move = player.choose_move(self.board)

        self.board.push(move)  # Make the move

    def is_game_over(self):
        return self.board.is_game_over()

    def __str__(self):

        column_labels = "\n----------------\na b c d e f g h\n"
        board_str =  str(self.board) + column_labels

        # did you know python had a ternary conditional operator?
        move_str = "White to move" if self.board.turn else "Black to move"

        return board_str + "\n" + move_str + "\n"
```

```python
import chess

class MinimaxAI():
    #Initialize MinMax with particular depth and node visited(default set to zero) 
    def __init__(self, depth):
        self.depth = depth
        self.nodes_visited = 0
    
    #Evaluation function to calcuate total sum of pieces on both sides for analyzing best possible next move
    #based on value
    def evaluation(self, state):
        total_pawns = 1 * abs(len(state.pieces(chess.PAWN, chess.WHITE)) - len(state.pieces(chess.PAWN, chess.BLACK)))
        total_knights = 3 * abs(len(state.pieces(chess.KNIGHT, chess.WHITE)) - len(state.pieces(chess.KNIGHT, chess.BLACK)))
        total_bishops = 3 * abs(len(state.pieces(chess.BISHOP, chess.WHITE)) - len(state.pieces(chess.BISHOP, chess.BLACK)))
        total_rooks = 5 * abs(len(state.pieces(chess.ROOK, chess.WHITE)) - len(state.pieces(chess.ROOK, chess.BLACK)))
        total_queens = 9 * abs(len(state.pieces(chess.QUEEN, chess.WHITE)) - len(state.pieces(chess.QUEEN, chess.BLACK)))
        total_king = 100 * abs(len(state.pieces(chess.KING, chess.WHITE)) - len(state.pieces(chess.KING, chess.BLACK)))
        evaluation_sum  = total_pawns + total_knights + total_bishops + total_rooks + total_queens + total_king
        return evaluation_sum

    #Cutoff function for terminating the game or break search algorithm for retrieving best possible next move
    def cutoff(self, state, max_depth):
        if max_depth == self.depth or state.is_game_over(): 
            return True
        return False

    #Choose best move from possible next moves based on MinMax function
    def choose_move(self, state):
        nodes_visited  = 0
        moves = list(state.legal_moves)

        #prioritize moves(helpful for pruning and achieving faster results)
        possible_next_moves = []
        for move in moves:
            if move in set(state.generate_legal_captures()):
                possible_next_moves.insert(0,move)
            else:
                possible_next_moves.append(move)

        #Achieving best possible next move for minimax algo
        cur_val = float('inf') if state.turn == chess.BLACK else float('-inf')
        best_possible_next_move = None
        for move in moves:
            state.push(move)
            val = self.minimaxfn(state,self.depth,nodes_visited)
            #Terminate condition for achieving result of game
            if state.is_game_over(): 
                print(state.outcome())
                if state.outcome().winner is None:
                    print("Stalemate")  
                else:             
                    print("Min-Max Wins")
                    print("Black Wins!" if state.turn == chess.WHITE else "White Wins!") 
                exit()
            state.pop()
            #Picking minimum value move or maximum value move based on current player
            if val <= cur_val and state.turn == chess.BLACK:
                best_possible_next_move = move
                cur_val = val
            elif val >= cur_val and state.turn == chess.WHITE:
                best_possible_next_move = move
                cur_val = val
        print("MinMax recommending move " + str(move))
        return best_possible_next_move

    
    def minimaxfn(self, state, depth, nodes_visited):
        # If either sides can't move or match ends in draw
        if state.is_stalemate(): 
            return nodes_visited
        # If king(any side) is attacked and unable to escape to another position or defended
        if state.is_checkmate():
            return 999999999
        #Maximizing Player turn
        if state.turn:
            self.nodes_visited += 1
            max_val = float('-inf') 
            possible_moves = list(state.legal_moves)
            if self.cutoff(state,depth): # Check for termination or comparing maxdepth with curdepth
                return self.evaluation(state)
            for move in possible_moves: #iterating over set of possible moves
                state.push(move)
                max_val = max(max_val,self.minimaxfn(state,depth+1,nodes_visited))
                state.pop()
            return max_val
        #Minimizing Player turn
        else:
            self.nodes_visited += 1
            min_val = float('inf')
            possible_moves = list(state.legal_moves)
            if self.cutoff(state,depth): # Check for termination or comparing maxdepth with curdepth
                return self.evaluation(state)
            for move in possible_moves: #iterating over set of possible moves
                state.push(move)
                min_val = min(min_val,self.minimaxfn(state,depth+1,nodes_visited))
            state.pop()
        return min_val

 
    
    
        
```
